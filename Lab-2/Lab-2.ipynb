```python
#Run this cell
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next

def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

**Task 1:** Alternate Merge

```python
def alternate_merge(head1, head2):
  #TO DO
  temp1=head1
  temp2=head2

  while temp1!=None and temp2!=None :
    store1=temp1.next
    store2=temp2.next
    temp1.next=temp2
    temp2.next=store1

    temp1=store1
    temp2=store2
  return head1

print('==============Test Case 1=============')
head1 = createList(np.array([1,2,6,8,11]))
head2 = createList(np.array([5,7,3,9,4]))

print("Linked List 1:")
printLinkedList(head1)
print("Linked List 2:")
printLinkedList(head2)

head = alternate_merge(head1, head2)
print("Merged Linked List:")
printLinkedList(head)    #This should print    1 --> 5 --> 2 --> 7 --> 6 --> 3 --> 8 --> 9 --> 11 --> 4


print('==============Test Case 2=============')
head1 = createList(np.array([5, 3, 2, -4]))
head2 = createList(np.array([-4, -6, 1]))

print("Linked List 1:")
printLinkedList(head1)
print("Linked List 2:")
printLinkedList(head2)

head = alternate_merge(head1, head2)
print("Merged Linked List:")
printLinkedList(head)    #This should print    5 → -4 -> 3 → -6 -> 2 -> 1 -> -4


print('==============Test Case 3=============')
head1 = createList(np.array([4, 2, -2, -4]))
head2 = createList(np.array([8, 6, 5, -3]))

print("Linked List 1:")
printLinkedList(head1)
print("Linked List 2:")
printLinkedList(head2)

head = alternate_merge(head1, head2)
print("Merged Linked List:")
printLinkedList(head)    #This should print   4-> 8 → 2-> 6 → -2 → 5 → -4 -> -3
```

**Task 3:** ID Generator

```python
def idGenerator(head1, head2, head3):
  #TO DO
  previous=None
  current=head1
  while current!=None:
    next_node=current.next
    current.next=previous
    previous=current
    current=next_node
  reversed_head1=previous

  temp1=head2
  temp2=head3
  dummy_head=Node(None)
  tail=dummy_head

  while temp1!=None and temp2!=None:
    sum=(temp1.elem+temp2.elem)
    if sum>=10:
      result=sum%10
      new_node=Node(result)
      tail.next=new_node
      tail=new_node
    elif sum<10:
      new_node=Node(sum)
      tail.next=new_node
      tail=new_node
    temp1=temp1.next
    temp2=temp2.next
  summed_list=dummy_head.next


  temp = reversed_head1
  while temp.next!=None:
    temp = temp.next
  temp.next = summed_list
  return reversed_head1






print('==============Test Case 1=============')
head1 = createList(np.array([0,3,2,2]))
head2 = createList(np.array([5,2,2,1]))
head3 = createList(np.array([4,3,2,1]))

print("Linked List 1:")
printLinkedList(head1)
print("Linked List 2:")
printLinkedList(head2)
print("Linked List 3:")
printLinkedList(head3)

result = idGenerator(head1, head2, head3)
print("New ID:")
printLinkedList(result)    #This should print  2 → 2 → 3 → 0 → 9 → 5 → 4 → 2


print('==============Test Case 2=============')
head1 = createList(np.array([0,3,9,1]))
head2 = createList(np.array([3,6,5,7]))
head3 = createList(np.array([2,4,3,8]))

print("Linked List 1:")
printLinkedList(head1)
print("Linked List 2:")
printLinkedList(head2)
print("Linked List 3:")
printLinkedList(head3)

result = idGenerator(head1, head2, head3)
print("New ID:")
printLinkedList(result)    #This should print 1 → 9 → 3 → 0 → 5 → 0 → 8 → 5
```

```python
class Patient:
  #write a constructor
  def __init__(self, id, name, age, bloodgroup, next=None, prev=None):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup
    self.next = next
    self.prev = prev
class DoublyNode:
	def __init__(self, elem=None, next=None, prev=None):
		self.elem = elem
		self.next = next
		self.prev = prev
```

```python
class DoublyNode:
    def __init__(self, elem=None, next=None, prev=None):
        self.elem = elem
        self.next = next
        self.prev = prev
class WRM:

  def __init__(self):
    #Creating the dummy head
    self.dh = Patient(None,None,None,None,None,None)
    self.dh.next = self.dh
    self.dh.prev = self.dh
    self.head = None
    self.tail = None
    self.len = 0


  def registerPatient(self,id, name, age, bloodgroup):
    #To Do
    new_patient = Patient(id, name, age, bloodgroup)
    new_node = DoublyNode(new_patient)

    if self.head is None:
      self.head = new_node
      self.tail = new_node
      self.head.next = self.dh
      self.head.prev = self.dh
      self.dh.next = self.head
      self.dh.prev = self.tail
    else:
      self.tail.next = new_node
      new_node.prev = self.tail
      new_node.next = self.dh
      self.dh.prev = new_node
      self.tail = new_node
    self.len += 1

  def servePatient(self):
    #To Do

        if self.head is None:
            print("No more patients to be served.")
            return

        served_patient = self.head.elem
        print(served_patient.name, "is served.")

        if self.head == self.tail:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = self.dh
            self.dh.next = self.head

        self.len -= 1

  def showAllPatient(self):
    #To Do
        if self.head is None:
            print("No more patients left.")
            return

        temp = self.head
        while temp != self.dh:
            print(temp.elem.name, end=" ")
            temp = temp.next
        print()

  def canDoctorGoHome(self):
    #To Do
        if self.head is None:
            print("Yes, the doctor can go home.")
        else:
            print("No, there are still patients waiting.")

  def cancelAll(self):
    #To Do
        if self.head is None:
            print("No appointments left to be cancelled.")
            return

        self.head = None
        self.tail = None
        self.len = 0
        print("All appointments cancelled.")

  def ReverseTheLine(self):

        if self.head is None or self.head == self.tail:
            print("Queue is empty or has only one patient. No need to reverse.")
            return

        current = self.head
        prev = None

        while current != self.dh:
            next_node = current.next
            current.next = prev
            current.prev = next_node
            prev = current
            current = next_node

        self.tail, self.head = self.head, self.tail
        self.tail.next = self.dh
        self.head.prev = self.dh
        self.dh.next = self.head
        self.dh.prev = self.tail


wrm = WRM()

while True:
    print("\nOptions:")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Reverse")
    print("7. Quit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")
        wrm.registerPatient(id, name, age, blood_group)

    elif choice == 2:
        wrm.servePatient()

    elif choice == 3:
        wrm.showAllPatient()

    elif choice == 4:
        wrm.canDoctorGoHome()

    elif choice == 5:
        wrm.cancelAll()

    elif choice == 6:
        wrm.ReverseTheLine()

    elif choice == 7:
        break

    else:
        print("Invalid choice, please try again.")

#Write a Tester Code in this cell
print("\n**Welcome to Waiting Room Management System**")
wrm = WRM()
wrm.registerPatient(1, "John", 22, "O")
wrm.registerPatient(2, "Alice", 30, "A")
wrm.registerPatient(3, "Bob", 25, "B")
wrm.registerPatient(4, "David", 35, "AB")
wrm.showAllPatient()
wrm.ReverseTheLine()
wrm.showAllPatient()
wrm.servePatient()
wrm.ReverseTheLine()
wrm.servePatient()
wrm.showAllPatient()
wrm.canDoctorGoHome()
wrm.cancelAll()
```
