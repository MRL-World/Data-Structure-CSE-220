```python
#Task 02: Decryption Process

def decrypt_matrix(matrix):
  import numpy as np
  #To Do

  row,col=matrix.shape
  new_array=np.zeros((col),dtype=int)
  sum=0
  for i in range(col):
    for j in range(row):
      sum+=matrix[j][i]
      new_array[i]=sum
    sum=0

  final_array=np.zeros((col-1),dtype=int)

  for k in range(len(new_array)-1):
    final_array[k]=new_array[k+1]-new_array[k]
  return final_array

#DO NOT CHANGE THE CODE BELOW
matrix=np.array([[1,3,1],
                 [6,4,2],
                 [5,1,7],
                 [9,3,3],
                 [8,5,4]
                 ])

returned_array=decrypt_matrix(matrix)
print(returned_array)
#This should print [-13, 1]
```

```python
#Task 04: Matrix Compression

def compress_matrix(mat):
  #TO DO

  row,col=mat.shape
  new_arr=np.zeros((row//2,col//2),dtype=int)
  for i in range(0,row,2):
    for j in range(0,col,2):
      new_arr[i//2][j//2]=mat[i][j]+mat[i+1][j]+mat[i][j+1]+mat[i+1][j+1]

  return new_arr



#DO NOT CHANGE THE CODE BELOW
matrix=np.array([[1,2,3,4],
                 [5,6,7,8],
                 [1,3,5,2],
                 [-2,0,6,-3]
                 ])
print_matrix(matrix)
print("...............")
returned_array=compress_matrix(matrix)
print_matrix(returned_array)
#This should print

#|  14  |  22 |
#--------------
#|  2  |  10  |
#--------------
```

```python
#Task 05: Game Arena

  #TO DO
def play_game(arena):
    import numpy as np
    toal_points = 0
    row, col = len(arena), len(arena[0])
    for i in range(row):
        for j in range(col):
            if arena[i][j] % 50 == 0 and arena[i][j] != 0:
                player_points = 0
                if i > 0 and arena[i-1][j] == 2:
                    player_points += 2
                if i < row - 1 and arena[i+1][j] == 2:
                    player_points += 2
                if j > 0 and arena[i][j-1] == 2:
                    player_points += 2
                if j < col - 1 and arena[i][j+1] == 2:
                    player_points += 2
                toal_points += player_points
    if toal_points >= 10:
        print(f"Points Gained: {toal_points}. Your team has survived the game.")
    else:
        print(f"Points Gained: {toal_points}. Your team is out.")

#DO NOT CHANGE THE CODE BELOW
arena=np.array([[0,2,2,0],
                [50,1,2,0],
                [2,2,2,0],
                [1,100,2,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 6. Your team is out.
print(".....................")
arena=np.array([[0,2,2,0,2],
                [1,50,2,1,100],
                [2,2,2,0,2],
                [0,200,2,0,0]
                ])
print_matrix(arena)
play_game(arena)
#This should print
#Points Gained: 14. Your team has survived the game.
```
