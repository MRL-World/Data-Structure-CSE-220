```python
class Node:
  def __init__(self,o_vertex,edge_weight):
    self.o_vertex=o_vertex
    self.edge_weight=edge_weight
    self.next=None
```

```python
def add_edge(graph,in_vertex,out_vertex,weight):
  new_node=Node(out_vertex,weight)
  new_node.next=graph[in_vertex]
  graph[in_vertex]=new_node
  return graph
```

```python
# Task:4 (Adjacancy matrix)

def create_adjacency_matrix():
  matrix = np.zeros((12, 12), dtype=int)

  matrix[0][1] = 2
  matrix[0][5] = 2
  matrix[0][9] = 4
  matrix[0][11] = 5

  matrix[1][6] = 8

  matrix[2][4] = 3
  matrix[2][3] = 5

  matrix[3][5] = 9
  matrix[3][11] = 6

  matrix[4][6] = 3
  matrix[4][11] = 9

  matrix[5][7] = 4

  matrix[6][10] = 7

  matrix[7][0] = 3
  matrix[7][8] = 4

  matrix[8][9] = 2

  matrix[9][10] = 7

  return matrix

def convert_undirected_weighted_matrix(matrix):
  length = len(matrix)
  undirected_matrix = np.zeros((length, length), dtype=int)

  for i in range(length):
    for j in range(length):
      if matrix[i][j] != 0:
        undirected_matrix[i][j] = matrix[i][j]
        undirected_matrix[j][i] = matrix[i][j]

  return undirected_matrix

def print_matrix(matrix):
  for i in range(len(matrix)):
    print(f"Vertex {i}:", end=" ")
    for j in range(len(matrix[i])):
      if matrix[i][j] != 0:
        print(f"-> (V:{j}, W:{matrix[i][j]})", end=" ")
    print()

# Driver Code
received_matrix = create_adjacency_matrix()
undirected_matrix = convert_undirected_weighted_matrix(received_matrix)

print("Undirected Graph (Adjacency Matrix):")
print_matrix(undirected_matrix)
```
