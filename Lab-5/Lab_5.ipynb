### **Task-2:**

```python
def task2A( arr,i=0 ):
  while i<len(arr):
    print(arr[i],end=" ")
    i+=1

def task2B_recursive( arr,i=0 ):
  if i==len(arr):
    return
  print(arr[i],end=" ")
  task2B_recursive(arr,i+1)

def task2C( arr,i=0 ):
  sum=0
  while i<len(arr):
    sum+=arr[i]
    i+=1
  return sum


def task2D_recursive( arr,i=0 ):
  if i==len(arr):
    return 0
  return arr[i]+task2D_recursive(arr,i+1)


def task2E( arr ):
  odd_multiplication=1
  even_sum=0
  for i in range(len(arr)):
    if arr[i]%2!=0:
      odd_multiplication*=arr[i]
    else:
      even_sum+=arr[i]
  subtracted_value=odd_multiplication-even_sum
  return subtracted_value


######################################################
def sum_even(arr,i=0):
  if i==len(arr):
    return 0
  a=sum_even(arr,i+1)

  if arr[i]%2==0:
    return arr[i]+a
  return a

def multiply_odd(arr,i=0):
  if i==len(arr):
    return 1
  b=multiply_odd(arr,i+1)

  if arr[i]%2!=0:
    return arr[i]*b
  return b

def task2F_recursive( arr ):
  odd_product=multiply_odd(arr)
  even_product=sum_even(arr)
  return odd_product-even_product

```

```python
def task4A_recursive( head ):
  previous=None
  current=head
  while current!=None:
    next_value=current.next
    current.next=previous
    previous=current
    current=next_value
  return previous


def task4B_recursive( head ):
  if head==None:
    return
  task4B_recursive(head.next)
  print(head.elem,end=" ")
```
